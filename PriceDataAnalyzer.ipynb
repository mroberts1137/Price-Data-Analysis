```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def SMA(x, l):
    x_len = len(x)
    x = np.pad(x, (l, 0), 'edge')
    window = np.zeros(l)
    mu = np.zeros(x_len)

    for i in range(x_len):
        window = x[i:i+l]
        mu[i] = np.ma.average(window)
    return mu
```

```python
df = pd.read_csv('SPY-daily-2022-2023.csv', header=0, usecols=['Close'])
df
```

```python
#x = np.random.rand(100)

mu = SMA(df, 10)


plt.plot(df)
plt.plot(mu)
plt.show()
```
